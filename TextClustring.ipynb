```python
from time import time
import numpy as np
import sklearn.feature_extraction.text as text
import lda
import pandas as pd
```

```python
n_samples = 4000
n_features = 4000
n_topics = 5
n_top_words = 30
```

```python
data = pd.read_csv('data\RedPillMorocco.csv')
data = data.dropna()
titles = tuple(data.values.tolist())
test = data['text'].values.tolist()
```

```python
# nltk arabic stop words

from nltk.corpus import stopwords
stop_words = set(stopwords.words('arabic'))
```

```python
tf_vectorizer = text.CountVectorizer(max_df=0.95, min_df=2, max_features=n_features, stop_words=stop_words)
tf = tf_vectorizer.fit_transform(test)
vocab = tf_vectorizer.get_feature_names()
```

```python
model = lda.LDA(n_topics=5, n_iter=1000, random_state=1)
model.fit(tf)
```

```python
vocab
```

```python
topic_word = model.topic_word_
n = 5
for i, topic_dist in enumerate(topic_word):
    topic_words = np.array(vocab)[np.argsort(topic_dist)][:-(n+1):-1]
    print('*Topic {}\n- {}'.format(i, ', '.join(topic_words)))
```

```python
doc_topic = model.doc_topic_
for n in range(40):
    topic_most_pr = doc_topic[n].argmax()
    print("\ndoc: {} topic: {}\n{}...".format(n,
                                            topic_most_pr,
                                            titles[n][:50]))
```
