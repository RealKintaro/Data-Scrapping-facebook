```python
# delete dics with no comments key
data = [item for item in data if "comments" in item.keys()]
```

```python
# write a list of dicts to a json file
import json

json = json.dumps(data,ensure_ascii = False,indent=2)
# write json to file
with open(f"data/{page_name}.json", "w", encoding="utf-8") as f:
    f.write(json)
```

```python
comments = []

for item in data:
    comments.append(item['comments'])
```
